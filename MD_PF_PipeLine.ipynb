```python
data = data.drop('Id', axis=1)
```

```python
# Separar X (variables predictoras) e Y (variable objetivo)
X = data.drop('SalePrice', axis=1)  # Todas las columnas EXCEPTO SalePrice
Y = data['SalePrice']                # Solo SalePrice (variable a predecir)

print(f'X tiene {X.shape[0]} filas y {X.shape[1]} columnas')
print(f'Y tiene {Y.shape[0]} valores')
```

```python
# SimpleImputer de sklearn tiene problemas con tipo 'category'
# Necesita convertirlas a 'object' antes del pipeline

var_a_convertir = [
    'MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
    'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
    'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
    'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
    'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
    'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
    'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
    'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType',
    'SaleCondition', 'LotConfig', 'BsmtCond'
]

for col in var_a_convertir:
    X[col] = X[col].astype('object')

print('Variables categóricas convertidas a object')
```

```python
# Si no existen YearBuilt_Category e YearRemodAdd_Category, crearlas
if 'YearBuilt_Category' not in X.columns:
    X['YearBuilt_Category'] = pd.cut(
        X['YearBuilt'],
        bins=[1872, 1920, 1950, 1980, 2010],
        labels=["1872-1920", "1921-1950", "1951-1980", "1981-2010"],
        right=True,
        include_lowest=True
    )

if 'YearRemodAdd_Category' not in X.columns:
    X['YearRemodAdd_Category'] = pd.cut(
        X['YearRemodAdd'],
        bins=[1950, 1980, 2010],
        labels=["1950-1980", "1981-2010"],
        right=True,
        include_lowest=True
    )

print('Columnas de años categóricos verificadas/creadas')
```

```python
# Verificar que todas las variables del modelo existan en X
variables_faltantes = [v for v in variables if v not in X.columns]

if variables_faltantes:
    print(f'Variables faltantes en X: {variables_faltantes}')
else:
    print(f'Todas las {len(variables)} variables esperadas están presentes')

print(f'\nX tiene {X.shape[1]} columnas (se esperaban {len(variables)})')
```

```python
# Función para eliminar atípicos (similar a la profesora)
def limpiar_atipicos(X):
    """
    Limpia valores atípicos reemplazándolos con np.nan
    """
    # Limpiar LotFrontage (valores >= 270)
    X.loc[X["LotFrontage"] >= 270, "LotFrontage"] = np.nan

    # Limpiar TotalBsmtSF (valores > 5500)
    X.loc[X["TotalBsmtSF"] > 5500, "TotalBsmtSF"] = np.nan

    # Limpiar BsmtFinSF1 (valores > 5500)
    X.loc[X["BsmtFinSF1"] > 5500, "BsmtFinSF1"] = np.nan

    return X

print('Función de limpieza de atípicos definida')
```

```python
# Transformaciones para variables NUMÉRICAS
var_numericas = ['LotFrontage', 'TotalBsmtSF', 'BsmtFinSF1', 'MasVnrArea']

num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))  # Imputa con la media
])

print('Transformer para variables numéricas definido')
```

```python
# Variables categóricas a procesar
var_categoricas = [
    'MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
    'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
    'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
    'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
    'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
    'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
    'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
    'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature', 'SaleType',
    'SaleCondition', 'YearBuilt_Category', 'YearRemodAdd_Category'
]

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(drop='if_binary',handle_unknown='ignore', sparse_output=False))
])


print('Transformer para variables categóricas definido')
```

```python
# ColumnTransformer para aplicar diferentes transformaciones a diferentes columnas
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, var_numericas),
        ('cat', cat_transformer, var_categoricas)
    ],
    remainder='drop'  # Mantiene el resto de columnas sin cambios
)

print('ColumnTransformer definido')
```

```python
# Pipeline que limpia atípicos y luego prepara los datos
pipeline_preprocesamiento = Pipeline(steps=[
    ('limpiar_atipicos', FunctionTransformer(limpiar_atipicos)),
    ('preprocessor', preprocessor)
])

print('Pipeline de preprocesamiento completo definido')
```

```python
# Aplicar el pipeline de preprocesamiento a los datos
X_procesado = pipeline_preprocesamiento.fit_transform(X)
print(f'Datos procesados: {X_procesado.shape}')
```

```python
# Reentrenar el modelo RandomForest con datos procesados
modeloRF.fit(X_procesado, Y)
print('✓ Modelo reentrenado con datos procesados')
```

```python
# Pipeline FINAL: Preprocesamiento + Modelo Entrenado
pipeline_final = Pipeline(steps=[
    ('preprocesamiento', pipeline_preprocesamiento),
    ('modelo', modeloRF)
])
print('✓ Pipeline final creado exitosamente')
```

```python
# Guardar el pipeline final completo
import pickle
filename_pipeline = 'pipeline_final.pkl'
with open(filename_pipeline, 'wb') as f:
    pickle.dump(pipeline_final, f)
print(f'✓ Pipeline guardado como: {filename_pipeline}')
```
